```python
from PIL import Image
from PIL import ImageDraw
```

```python
for i in range(1, 6):
  img1=ROOT_DIR +"/images"+"/"+str(i)+".jpg"
 
  # creating a object
  im = Image.open("/content/drive/MyDrive/Data Visualization/images/"+str(i)+".jpg")

  json1=ROOT_DIR +"/data"+"/"+str(i)+".json"

  f = open(json1)
 
  # returns JSON object as
  # a dictionary
  data = json.load(f)
  l=int(im.size[0])/100
  m=int(im.size[1])/100
  a = []
  dack = Image.new('RGBA', im.size)
  dack.paste(im)
  poly = Image.new('RGBA', (1500,1500))

  for j in range(1,len(data)-1):
    a=list(tuple(map(tuple,data[j]['value']['points'])))
    a.clear()
    minx=100000
    miny=100000
    maxx=0
    maxy=0
    for i,k in data[j]['value']['points']:
      x=i*l
      y=k*m
      a.append([x,y])
      minx=min(minx,x)
      miny=min(miny,y)
      maxx=max(maxx,x)
      maxy=max(maxy,y)
    
    b=list(tuple(map(tuple,a)))
    from PIL import Image
    from PIL import ImageDraw


    


    poly = Image.new('RGBA', (1500,1500))

    
    
    
    pdraw = ImageDraw.Draw(poly)
    pdraw.polygon(b,fill=(255,255-j*20,150+j*20,127),outline=(255,255,255,255))

    dack.paste(poly, (0,0), mask=poly)
    
    
    
    from PIL import Image
    from PIL import ImageFont
    from PIL import ImageDraw

    # image opening
    

    # creating a copy of original image
    

    # Image is converted into editable form using
    # Draw function and assigned to draw
    draw = ImageDraw.Draw(dack)

    # ("font type",font size)


    # Decide the text location, color and font
    # (255,255,255)-White color text
    draw.text((minx, miny), data[j]['value']['polygonlabels'][0], (0, 0, 0))
    draw.rectangle(((minx,miny), (maxx, maxy)))

    dack.show()

  dack
```
